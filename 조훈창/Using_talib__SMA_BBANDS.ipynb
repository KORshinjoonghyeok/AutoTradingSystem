```python
diff = df['close'] - df['open']
df['diff'] = diff
df['color'] = 'blue'
```

```python
def get_color(df_):
    dif = df_['diff']
    for i in range(len(df_)):
        if dif.iloc[i] >= 0:
            df_['color'].iloc[i] = 'red'
        else:
            df_['color'].iloc[i]='blue'
```

```python
get_color(df)
df
```

```python
df['date'] = df['date'].astype('str')
df = df.iloc[:100,:]
df.set_index(df['date'],inplace=True)
```

```python
df.sort_index(ascending=True,inplace=True)
```

```python
df
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df['SMA'] = ta.SMA(df['close'])
```

```python
df['BBAND_UPPER'],df['BBAND_MIDDLE'],df['BBAND_LOWER'] = ta.BBANDS(df['close'],20,2)
```

```python
df['MOM_5'] = ta.MOM(df['close'],timeperiod=5)
```

```python
plt.figure(figsize = (16,8))
plt.bar(df['date'],height=df['diff'],width = 0.8,bottom=df['open'],color = df['color'])
plt.plot(df['date'],df['BBAND_UPPER'])
plt.plot(df['date'],df['BBAND_MIDDLE'])
plt.plot(df['date'],df['BBAND_LOWER'])
plt.xticks([df['date'].iloc[0],df['date'].iloc[-1]])
plt.show()
```
